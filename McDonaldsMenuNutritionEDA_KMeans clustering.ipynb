```python
from scipy import stats
from sklearn.cluster import KMeans
```

```python
X_cluster = df[['Calories', 'Saturated Fat (g)', 'Trans Fat (g)',
                'Cholesterol (g)', 'Sodium (g)', 'Carbs (g)', 'Fiber (g)', 'Sugars (g)', 'Protein (g)']]
num_clusters= 3
kmeans=KMeans(n_clusters=num_clusters,random_state=42)
```

```python
clusters = kmeans.fit_predict(X_cluster)
```

```python
df['Cluster']=clusters
```

```python
plt.scatter(df['Calories'], df['Sodium (g)'], c=clusters, cmap='viridis')
plt.xlabel('Calories')
plt.ylabel('Sodium (g)')
plt.show()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from gensim import corpora, models
```

```python
nltk.download('punkt')
```

```python
df['Item'] = df['Item'].apply(lambda x: word_tokenize(x.lower()))
```

```python
dictionary = corpora.Dictionary(df['Item'])
corpus = [dictionary.doc2bow(item) for item in df['Item']]
```

```python
lda_model = models.LdaModel(corpus, num_topics=2, id2word=dictionary, passes=15)
```

```python
topics = lda_model.print_topics(num_words=3)
for topic in topics:
    print(topic)
```
